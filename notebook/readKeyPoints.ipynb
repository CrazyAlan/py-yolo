```python
def keypoints_2_box(keys):
    xs = filter(lambda x: x != 0, keys[0::3])
    if len(xs) == 0:
        xs = [0]
    ys = filter(lambda x: x != 0, keys[1::3])
    if len(ys) == 0:
        ys = [0]
    x_min, x_max = np.min(xs), np.max(xs)
    y_min, y_max = np.min(ys), np.max(ys)
    return [int(x_min), int(y_min), int(x_max), int(y_max)]
```
